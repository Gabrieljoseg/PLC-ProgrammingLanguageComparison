### Como funciona Flutter

#### O Flutter é dividido em 3 camadas:

O Flutter é uma plataforma de desenvolvimento de aplicativos móveis que consiste em três camadas principais:

- **Framework Flutter:** é a camada responsável por fornecer as ferramentas e componentes de interface do usuário para desenvolver aplicativos móveis. O framework Flutter é escrito em Dart e contém vários pacotes, como material, cupertino e widgets, que fornecem componentes de interface do usuário pré-construídos para desenvolvedores de aplicativos.

- **Engine Flutter:** é a camada de baixo nível da plataforma Flutter, responsável pela renderização da interface do usuário e pela comunicação com o sistema operacional subjacente. A engine Flutter é escrita em C/C++ e fornece uma plataforma de desenvolvimento de aplicativos rápida e responsiva.

- **Embedder Flutter:** é a camada que liga o código Flutter ao sistema operacional subjacente. A camada Embedder Flutter é específica para cada plataforma (iOS, Android, Windows, Linux, etc.) e fornece uma interface para a engine Flutter, permitindo que o Flutter interaja diretamente com o sistema operacional e os recursos do dispositivo. Isso permite que o Flutter renderize a interface do usuário de forma nativa, independentemente da plataforma subjacente.

O uso dessas três camadas permite que os desenvolvedores criem aplicativos móveis de alta qualidade e desempenho, com uma única base de código que pode ser executada em várias plataformas. O Flutter é uma plataforma de desenvolvimento de aplicativos móveis popular por sua facilidade de uso, rapidez de desenvolvimento e capacidade de criar aplicativos nativos de alta qualidade com uma única base de código.

> **Widgets**: O Flutter é baseado em widgets. Widgets são elementos visuais básicos que podem ser combinados para formar interfaces de usuário completas. Existem widgets para praticamente tudo o que você precisa em uma interface de usuário, desde botões simples até tabelas complexas.

> **Hot Reload:** O Flutter possui uma funcionalidade de Hot Reload, que permite aos desenvolvedores visualizar rapidamente as alterações em seus códigos sem precisar reconstruir o aplicativo inteiro. Isso ajuda a acelerar o processo de desenvolvimento e torna a depuração mais fácil.

> **Material Design e Cupertino:** O Flutter inclui dois conjuntos de widgets de interface de usuário: Material Design, que é baseado no design do Google, e Cupertino, que é baseado no design da Apple. Isso permite que os desenvolvedores criem interfaces de usuário que sejam consistentes com o estilo visual de cada plataforma.

> **Acesso à plataforma:** O Flutter permite acesso total às APIs nativas da plataforma, o que significa que os desenvolvedores podem usar recursos específicos da plataforma, como câmera, GPS e notificações. Isso permite que os aplicativos criados com Flutter aproveitem totalmente as capacidades do dispositivo em que estão sendo executados.

> **Performance:** O Flutter é conhecido por ter um desempenho excelente, pois usa sua própria renderização de widget e o motor gráfico Skia. Isso permite que os aplicativos criados com Flutter sejam mais rápidos e suaves do que os criados com outros frameworks.

Em resumo, o Flutter é uma ferramenta poderosa para desenvolvimento de aplicativos móveis, que se baseia em widgets, oferece hot reload, inclui dois conjuntos de widgets de interface de usuário e permite acesso total às APIs nativas da plataforma, além de oferecer excelente desempenho.

## Dependências 

> Em um projeto Flutter, **as dependências são bibliotecas externas** que são importadas e usadas no código para adicionar funcionalidades específicas ao projeto. As dependências podem ser divididas em dois tipos:

- **Dependências do sistema:** são bibliotecas que já fazem parte do próprio Flutter SDK e podem ser usadas diretamente no projeto sem precisar de instalação adicional. Alguns exemplos são o dart:async, dart:math, dart:convert, entre outros.

- **Dependências externas:** são bibliotecas desenvolvidas por terceiros que oferecem funcionalidades específicas que não estão disponíveis no Flutter SDK. Essas dependências podem ser incluídas no projeto através de gerenciadores de pacotes, como o pub (o gerenciador de pacotes do Flutter) ou o gradle (para projetos que utilizam o Flutter em conjunto com o Android nativo).

> As dependências externas são especificadas no arquivo pubspec.yaml do projeto, que é o arquivo de configuração que descreve o nome, a versão e as dependências do projeto. Quando uma nova dependência é adicionada, é necessário executar o comando flutter packages get para baixar e instalar as dependências no projeto.

> As dependências podem ser organizadas em diferentes grupos, como dependencies (para as dependências principais do projeto), dev_dependencies (para dependências usadas apenas durante o desenvolvimento, como ferramentas de teste) e dependency_overrides (para substituir as dependências padrão do Flutter SDK por outras versões).

> No geral, o uso de dependências é uma prática comum em projetos Flutter, pois permite que os desenvolvedores adicionem funcionalidades adicionais ao projeto sem precisar reinventar a roda ou escrever todo o código do zero.

### Sync and aSync

> Em programação, sync (síncrono) e async (assíncrono) são dois modos de execução de um programa ou trecho de código.

> No modo **síncrono**, o programa segue uma ordem de execução linear, ou seja, cada linha de código é executada em sequência. O programa aguarda a conclusão de cada tarefa antes de prosseguir para a próxima. Isso significa que o programa só pode realizar uma tarefa por vez, o que pode ser problemático em algumas situações, como por exemplo, em operações que podem levar muito tempo para serem concluídas, como fazer uma requisição à internet.

> No modo **assíncrono**, o programa não espera a conclusão de uma tarefa antes de prosseguir para a próxima. Isso significa que o programa pode iniciar várias tarefas ao mesmo tempo, o que pode tornar o processo de execução mais rápido e eficiente. No entanto, em alguns casos, o resultado das tarefas assíncronas não estará disponível imediatamente, o que pode complicar o fluxo de execução do programa.

Em linguagens de programação como Dart, que é a linguagem usada no desenvolvimento de aplicativos Flutter, os modos sync e async são implementados usando a programação assíncrona. Em Dart, um método assíncrono é marcado com a palavra-chave async, e pode conter um ou mais pontos de espera (await), que indicam que o método deve aguardar a conclusão de uma tarefa assíncrona antes de prosseguir.

Por exemplo, ao fazer uma requisição à internet em um aplicativo Flutter, é necessário usar uma chamada assíncrona para não bloquear a thread principal do aplicativo. Nesse caso, pode-se usar a biblioteca http e um método assíncrono como o seguinte:



`````
import 'package:http/http.dart' as http;

Future<String> fetchAlbum() async {
  final response =
      await http.get(Uri.parse('https://jsonplaceholder.typicode.com/albums/1'));

  if (response.statusCode == 200) {
    return response.body;
  } else {
    throw Exception('Failed to load album');
  }
}

`````
Nesse exemplo, o método **fetchAlbum** é assíncrono (marcado com a palavra-chave **async**) e faz uma requisição à internet usando a biblioteca **http**. A chamada **await http.get()** é assíncrona, e faz com que o programa aguarde a conclusão da requisição antes de prosseguir para a próxima linha de código. Isso permite que o aplicativo continue respondendo aos usuários enquanto a requisição está em andamento.

### Estrutura de repetição

**FOR** 

``````python
// Exemplo de for para repetir uma ação 5 vezes:
for (int i = 0; i < 5; i++) {
  print("Ação executada $i vezes");
}

// Exemplo de for para iterar sobre uma lista:
List<String> frutas = ['maçã', 'banana', 'laranja'];
for (String fruta in frutas) {
  print("Gosto de comer $fruta");
}

``````

**WHILE**

``````python
// Exemplo de while para repetir uma ação enquanto um contador é menor que 5:
int contador = 0;
while (contador < 5) {
  print("Ação executada $contador vezes");
  contador++;
}

// Exemplo de while para repetir uma ação enquanto o usuário não digitar "sair":
String input = '';
while (input != 'sair') {
  print("Digite 'sair' para encerrar o programa");
  input = stdin.readLineSync()!;
}
``````

**DO...WHILE :** usado para repetir uma ação pelo menos uma vez, e depois repetir enquanto uma condição for verdadeira.

``````python
// Exemplo de do...while para repetir uma ação até que o usuário digite "sair":
String input;
do {
  print("Digite 'sair' para encerrar o programa");
  input = stdin.readLineSync()!;
} while (input != 'sair');
``````

**forEach:** usado para iterar sobre uma coleção sem precisar de um índice.


``````python
// Exemplo de forEach para iterar sobre uma lista de frutas:
List<String> frutas = ['maçã', 'banana', 'laranja'];
frutas.forEach((fruta) {
  print("Gosto de comer $fruta");
});
``````

**for...in:** usado para iterar sobre uma coleção de valores.

``````python
// Exemplo de for...in para iterar sobre uma lista de números e somá-los:
List<int> numeros = [1, 2, 3, 4, 5];
int soma = 0;
for (int numero in numeros) {
  soma += numero;
}
print("A soma dos números é $soma");

``````

**for...in com if:** podemos usar o for...in em conjunto com o if para filtrar os valores que serão iterados.

``````python
// Exemplo de for...in com if para iterar sobre uma lista de números pares:
List<int> numeros = [1, 2, 3, 4, 5, 6];
for (int numero in numeros) {
  if (numero % 2 == 0) {
    print("$numero é par");
  }
}
``````

**break e continue:** podemos usar essas palavras-chave para interromper ou pular uma iteração do loop.

``````python
// Exemplo de break para interromper o loop quando encontrar um número ímpar:
List<int> numeros = [2, 4, 5, 6, 8, 9, 10];
for (int numero in numeros) {
  if (numero % 2 != 0) {
    print("Encontrou um número ímpar, encerrando o loop");
    break;
  }
  print("$numero é par");
}
``````

``````python
// Exemplo de continue para pular a iteração quando encontrar um número ímpar:
List<int> numeros = [1, 2, 3, 4, 5];
for (int numero in numeros) {
  if (numero % 2 != 0) {
    print("Encontrou um número ímpar, pulando para o próximo");
    continue;
  }
  print("$numero é par");
}
``````

**do...while com break:** podemos usar o do...while em conjunto com o break para interromper o loop quando uma condição for satisfeita.

``````python
// Exemplo de do...while com break para repetir uma ação até que o usuário digite "sair":
String input;
do {
  print("Digite 'sair' para encerrar o programa");
  input = stdin.readLineSync()!;
  if (input == 'erro') {
    print("Comando inválido, encerrando o programa");
    break;
  }
} while (input != 'sair');
``````

###Funções

Para definir uma função em Dart, utilizamos a palavra-chave "void" se a função não retorna nada, ou o tipo de retorno da função se ela retorna um valor.

``````python
void helloWorld() {
  print("Hello, world!");
}

int soma(int a, int b) {
  return a + b;
}
``````

A função helloWorld() não recebe parâmetros e não retorna nenhum valor, apenas imprime a mensagem "Hello, world!" na tela.

A função soma(int a, int b) recebe dois parâmetros 'a' e 'b', ambos do tipo 'int', e retorna a soma desses dois valores.

``````python
void main() {
  helloWorld();
  
  int resultado = soma(3, 5);
  print("O resultado da soma é $resultado");
}
``````

No exemplo acima, chamamos a função 'helloWorld()' na primeira linha do método 'main()', e depois chamamos a função 'soma(3, 5)' para obter a soma dos números 3 e 5. O 'resultado' é armazenado na variável resultado e depois impresso na tela.

``````python
void main() {
  List<int> numeros = [1, 2, 3, 4, 5];

  numeros.forEach((numero) {
    print(numero * 2);
  });
}
``````

No exemplo acima, utilizamos o método 'forEach()' da lista 'numeros' para executar uma função anônima para cada elemento da lista. A função anônima recebe um parâmetro 'numero' e imprime o dobro desse número na tela.

## Problemas simples com Dart

#### 1.Crie um programa que receba dois números e imprima a soma deles.

``````python
import 'dart:io';

void main() {
  print("Digite o primeiro número:");
  int num1 = int.parse(stdin.readLineSync()!);

  print("Digite o segundo número:");
  int num2 = int.parse(stdin.readLineSync()!);

  int resultado = num1 + num2;
  print("A soma de $num1 e $num2 é $resultado");
}
``````

#### 2.Crie um programa que receba um número e imprima se ele é par ou ímpar.

``````python
import 'dart:io';

void main() {
  print("Digite um número:");
  int numero = int.parse(stdin.readLineSync()!);

  if (numero % 2 == 0) {
    print("$numero é par");
  } else {
    print("$numero é ímpar");
  }
}
``````

#### 3.Crie um programa que receba uma lista de números e imprima o maior e o menor valor.

``````python
import 'dart:io';

void main() {
  print("Digite uma lista de números separados por espaço:");
  List<String> numerosTexto = stdin.readLineSync()!.split(" ");
  List<int> numeros = numerosTexto.map(int.parse).toList();

  int maior = numeros[0];
  int menor = numeros[0];

  for (int i = 1; i < numeros.length; i++) {
    if (numeros[i] > maior) {
      maior = numeros[i];
    }
    if (numeros[i] < menor) {
      menor = numeros[i];
    }
  }

  print("O maior número é $maior e o menor número é $menor");
}
``````

#### 4.Crie um programa que receba uma lista de nomes e imprima todos os nomes que começam com a letra "A".

``````python
import 'dart:io';

void main() {
  print("Digite uma lista de nomes separados por espaço:");
  List<String> nomes = stdin.readLineSync()!.split(" ");

  List<String> nomesComA = [];

  for (String nome in nomes) {
    if (nome[0].toUpperCase() == "A") {
      nomesComA.add(nome);
    }
  }

  print("Nomes que começam com a letra A: ${nomesComA.join(", ")}");
}
``````
